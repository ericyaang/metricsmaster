```python
import pandas as pd
import pymongo
import streamlit as st
from configparser import ConfigParser

# get credentials
parser = ConfigParser()
_ = parser.read('collect_data.cfg')
mangodb_cred = parser.get('mangodb', 'cred')
```

## Setup connection with MongoDB

See the [docs](https://pymongo.readthedocs.io)

```python
client = pymongo.MongoClient(mangodb_cred)
db = client['sample_airbnb']
collection = db['listingsAndReviews']
```

```python
dblist = client.list_database_names()
dblist
```

```python
collist = db.list_collection_names()
collist
```

```python
# Retrieve data from the MongoDB collection
cursor = collection.find()

# Convert the results to a DataFrame format
df = pd.DataFrame(list(cursor))
```
